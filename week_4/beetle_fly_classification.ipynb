```python
def get_label(ts, X_train, y_train):
    dtw_values = np.array([fastdtw.fastdtw(ts, ts_reference)[0] for ts_reference in X_train])
    np.where(dtw_values == min(dtw_values))
    idx = np.where(dtw_values == min(dtw_values))[0][0]
    label = y_train[idx]
    return label
```
